```python
# La libreria selenium automatizara la entrada de la pagina web de cada gen 
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
import openpyxl
import xlsxwriter

# Opciones de navegaci√≥n
options =  webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')

#la ruta driver_path cambiara en donde tenga instalado el driver , en mi caso es este. Agradeceria que lo descargue profesora lo puede encontrar en internet
driver_path = 'C:\\Users\\ALEX\\Desktop\\driver\\chromedriver.exe'
driver = webdriver.Chrome(driver_path, chrome_options=options)

#el tiempo de espera que le daremos a clickear
wait = WebDriverWait(driver, 5)

#Crearemos un archivo excel 
archivo=xlsxwriter.Workbook("Prueba.xlsx")
hoja=archivo.add_worksheet()

#Escribimos las cabezeras del archivo excel 

hoja.write(0,0,'Protein')
hoja.write(0,1,'Gene')
hoja.write(0,2,'Organism')
hoja.write(0,3,'Status')

def breast(gen):
    archivo=xlsxwriter.Workbook("Prueba.xlsx")
    
    #Pagina donde ingresara cada gen
    pag='https://www.uniprot.org/uniprot/?query='+gen+"+Capra+Hircus&sort=score"
    driver.get(pag)
    
    #Empezamos la automizacion
    
    WebDriverWait(driver, 10)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                      '/html/body/main/div/div[3]/div[2]/form/table/tbody/tr/td[2]/a')))\
    .click()
    
    #Extraemos informacion
    
    Protein = driver.find_element_by_xpath('//*[@id="content-protein"]').text
    Gene = driver.find_element_by_xpath('//*[@id="content-gene"]/h2').text
    Organism = driver.find_element_by_xpath('//*[@id="content-organism"]/em').text
    Status = driver.find_element_by_xpath('//*[@id="content-status"]/a/span').text
    
    #Escribimos en cada celda, el write lo hemos comenzado desde el 2 y avanzara uno en uno en cada celda
    hoja.write(write,0,Protein)
    hoja.write(write,1,Gene)
    hoja.write(write,2,Organism)
    hoja.write(write,3,Status)
    
    

    return(Protein,Gene,Organism,Status)


# Leeremos el excel genes para sacar informacion de cada hoja y de cada gen
# Hacemos una funcion para cada hoja en donde es editable el count<5 , puse 5 para que no corra todos los genes
# Si se desea ver la informacion de todos los genes cambiar el count<5 y poner el valor de iloc
# Por ejemolo en la primera hoja envez de poner count<5,pondremos count<212
# Es 212 ya que comenzamos del 1 por eso son 212 valores de la lista
# Dejaremos en 5 para que sea mas intutivo ,ya depende de usted cuantos valores desea leer.
# Recordar tambien que si no se encuentra ningun valor de un gen , esto lo omitira con las funciones try y except

df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C1 Milagros')
a=df.iloc[1:213, 4]
genes=list(a)
count=0
write=2
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C2 Wilmer')
a=df.iloc[1:325, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C3 Gisell')
a=df.iloc[1:579, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C4 Jerica')
a=df.iloc[1:318, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C5 Manuel')
a=df.iloc[1:726, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C7 Mariana')
a=df.iloc[1:649, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C8 Rosha')
a=df.iloc[1:270, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C9 Jerica')
a=df.iloc[1:118, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C10 Lucero')
a=df.iloc[1:262, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C11 Nicole')
a=df.iloc[1:465, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C12 Gisell')
a=df.iloc[1:135, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C13 Lucero')
a=df.iloc[1:454, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1


df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C14 Wilmer')
a=df.iloc[1:156, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1
    

df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C15Milagros')
a=df.iloc[1:265, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1



df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C14 Wilmer')
a=df.iloc[1:156, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1



df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C16')
a=df.iloc[1:270, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1



df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C17')
a=df.iloc[1:183, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1



df = pd.read_excel('02 Ch Genes ordenados con vecinos.xlsx',sheet_name='C18')
a=df.iloc[1:684, 4]
genes=list(a)
count=0
for i in genes:
    if count<5:
        try:            
            print(breast(i))
            write+=1
        except:
            "Holi"
    else:
        break
    count+=1

archivo.close()

# En los resultados se puede visualizar los elementos que se agregara a la tabla excel
```
